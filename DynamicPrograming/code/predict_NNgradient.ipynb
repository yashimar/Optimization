```python
class QuadraticDataset(data.Dataset):
    
    def __init__(self, length=100, width=1, error_range=0, x_divide=1.0, y_divide=1.0, seed=1234):
        torch.manual_seed(seed)
        np.random.seed(seed)
        random.seed(seed)
        
        self.length = length
        self.input = torch.empty(length, 1).uniform_(-width,width)
        self.output = self.input**2 + torch.empty(length, 1).uniform_(-error_range,error_range)
        self.input /= x_divide
        self.output /= y_divide
        
    def __len__(self):
        return self.length
    
    def __getitem__(self, index):
        x = self.input[index]
        y = self.output[index]
        
        return x, y
```

```python
dataset = QuadraticDataset(length=300,width=1.5, x_divide=1., y_divide=1., error_range=0, seed=42)
dataloader = torch.utils.data.DataLoader(dataset,batch_size=300,shuffle=True)
x, y = next(iter(dataloader))

plt.scatter(x,y)
```
